# P0 Fake data 

```python
import numpy as np
from scipy import stats

t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

# Problem 1: subsets of data

# A. Scatter plot of the first 1000 data points

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

x=t[:1000]
y=bkgd[:1000]
plt.scatter(x, y, marker='o');
plt.xlabel('time')
plt.ylabel('background level')
```

# B. Plotting some middle section of the data

2000 points starting as 500

```python
x=t[500:2500]
y=bkgd[500:2500]
plt.scatter(x, y, marker='o');
plt.xlabel('time')
plt.ylabel('background level')
```

# C. Plot every 100th point

```python
x=t[0:len(t):100]
y=bkgd[0:len(bkgd):100]
plt.scatter(x, y, marker='o');
plt.xlabel('time')
plt.ylabel('background level')
```

# Problem 2: density

# A. Make a 2d histogram plot with colorbar

```python
import matplotlib.pyplot as plt

plt.hist2d(t, bkgd, bins=100, range=None, density=False, weights=None, cmin=None, cmax=None)
plt.xlabel('time')
plt.ylabel('background level')
```

# B. Clearly explain what is being plotted in your plot (e.g. what does each color mean?)

The brighter the color, the more likely we find a background data point around this position(time, intensity)

# Problem 3: folding

# a. Make a new time variable that folds at a period of 2𝜋 then make a scatter plot of all your data

```python
import math

T=np.array(t)
pT=T % (2*math.pi)
```

```python
x=pT
y=bkgd
plt.scatter(x, y, marker='o');
plt.xlabel('time')
plt.ylabel('background level')
```

# b. Make a 2D histogram plot of your folded data

```python
plt.hist2d(pT, bkgd, bins=100, range=None, density=False, weights=None, cmin=None, cmax=None)
plt.xlabel('time')
plt.ylabel('background level')
```

# c. Average as a function of the folded variable

```python
import numpy as np

subarrays=np.array_split(bkgd, 1000)
avgbkgd = [np.mean(arr) for arr in subarrays]

subarraysT=np.array_split(pT, 1000)
avgT = [np.mean(arr) for arr in subarraysT]


plt.hist2d(avgT, avgbkgd, bins=100, range=None, density=False, weights=None, cmin=None, cmax=None)
plt.xlabel('time')
plt.ylabel('avg background level')
```
